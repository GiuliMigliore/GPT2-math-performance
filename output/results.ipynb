### Visualize the Outputs

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(data=result_flexible, x="beam", y="Value", hue="temp",
             marker="o", palette="viridis")

plt.xlabel("Beam Size")
plt.ylabel("Flexible Match Metric")
plt.title("Impact of Temperature and Beam Size on GPT-2 Performance on GSM8K Dataset")
plt.legend(title="Temperature")

plt.show()
plt. savefig('plot.png')
```
